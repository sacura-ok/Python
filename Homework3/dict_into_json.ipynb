```python
data_dict = {
    'Name': 'Paul',
    'Age': 25,
    'Student': False,
    'StudentID': None,
    'Works': True,
    'Salary': 135.35,
    'books': {'book1': 1998, 'book2': {'book1_1': 2001, 'book1_2': 2005}, 'book3': 2011},
    'CDs': ['CD1', 'CD2', 'CD3'],
}

# print(data_dict)


def type_of_value(value):
    return type(value)


def type_other(type_val, data):
    q = '"'
    if type_val == str:
        return f"{q}{data}{q}"
    elif type_val == int or type_val == float:
        return f"{data}"
    elif data == False or data == True:
        return f"{ str(data).lower()}"


def type_list_tuple(data):
    s = ''
    i = 0
    for value in data:
        s = s + f"{type_data(s, value)}"
        if i < len(data) - 1:
            s = s + f"{','}{' '}"
        i = i + 1
    return f"{'['}{s}{']'}"


def type_dict(s, data):
    q = '"'
    i = 0
    for key, value in data.items():
        s = s + f"{q}{key}{q}:{' '}{type_data(s, value)}"
        if i < len(data) - 1:
            s = s + f"{','}{' '}"
        i = i + 1
    return f"{'{'}{s}{'}'}"


def type_data(s, data):
    type_val = type_of_value(data)

    if type_val == dict:
        s = ''
        return type_dict(s, data)

    elif type_val == list or type_val == tuple:
        return type_list_tuple(data)

    elif type_val in [str, int, float, bool]:
        return type_other(type_val, data)

    elif data is None:
        return f"{'null'}"


def logger(filename):
    def decorator(func):
        def wrapped(*args, **kwargs):
            result = func(*args, **kwargs)
            with open(filename, 'w') as f:
               f.write(str(result))
            return result
        return wrapped
    return decorator


@logger('data.json')
def dict_json(data):
    s = ''
    return type_data(s,data)


dict_json(data_dict)
#print(dict_json(data_dict))
```
